```python
from bs4 import BeautifulSoup
```

```python
with open('index.html','r') as f:
    doc = BeautifulSoup(f,"html.parser")
```

```python
print(doc)
```

```python
print(doc.prettify())
```

```python
#fetches first identity that comes under this tag
tag = doc.title
print(tag)
```

```python
#fetches inside of the tag
print(tag.string)
```

```python
#modifying the tag
tag.string = 'hello'
```

```python
print(doc)
```

```python
#to fetch more than first occuring of tags
tags = doc.find_all('p')[0]
tags
```

```python
#to access the inside of the p tags, we use the same method to look inside p tags this time instead
print(tags.find_all("b"))
```

```python
import requests 
```

```python
url = "https://pk.sapphireonline.pk/collections/rtw-shirts/products/003PPRW25V14_999.html?source=plp"
```

```python
result = requests.get(url)
```

```python
doc  = BeautifulSoup(result.text,"html.parser")
```

```python
print(doc.prettify())
```

```python
prices = doc.find_all(string=lambda text:"Rs." in text if text else False)
for price in prices:
    print(price.strip())
```
